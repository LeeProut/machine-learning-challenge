# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("Resources/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

```python
# confirming null values have been dropped
df.isna().sum()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]

# selected_features = df[['koi_period', 'koi_impact', 'koi_depth', 'koi_insol', 'koi_slogg']]

# changing approach to choosing all features by just dropping the koi_disposition column
# if run_all_features == False:
#     selected_features = df[['koi_period', 'koi_impact', 'koi_depth']]
# else:
selected_features = df.drop("koi_disposition", axis=1)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
# sample code for reshaping:
# y = df['koi_disposition'].values.reshape(-1, 1)
# experiment with no reshaping
y = df['koi_disposition']
print(X.shape, y.shape)
```

```python
# format the FALSE POSITIVE value 
mask = df["koi_disposition"] == "FALSE POSITIVE"
df.loc[mask, "koi_disposition"] = "FALSE_POSITIVE"
df["koi_disposition"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from tensorflow.keras.utils import to_categorical
```

```python
# scaler = MinMaxScaler()
# standard scaler is st. deviation, MinMax scaler is -1, 1
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Create Model Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500)
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_test_scaled, y_test)
```

```python
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

## Feature Selection

```python
# view feature importances
sorted(zip(rf.feature_importances_, selected_features), reverse=True)
```

```python
# create dataframe for feature importances
tuned_features = pd.DataFrame({
    "Features": X.columns, 
    "Importance": rf.feature_importances_
})

tuned_features_sorted = tuned_features.sort_values("Importance", ascending=False)
tuned_features_sorted
```

```python
# select features importances of 0.012 and higher
tuned_features_selected = tuned_features_sorted.loc[tuned_features_sorted["Importance"] > 0.012]

tuned_features_selected
rf_features = tuned_features_selected[["Features"]]
rf_features_list = list(rf_features["Features"])
rf_features_list

# filter original df for the list of rf_features
df_rf_select = df.filter(rf_features_list)
df_rf_select
```

# Train the Model



```python
# trying rf_features in the model
X = df_rf_select
y = df['koi_disposition']
print(X.shape, y.shape)
```

```python
# Test, train, split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# scaler = MinMaxScaler()
# standard scaler is st. deviation, MinMax scaler is -1, 1
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# NEW random forest classifier with rf features selected
rf = RandomForestClassifier(n_estimators=500)
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_test_scaled, y_test)
```

```python
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# example for Random Forest Grid Search: 
# https://jamesrledoux.com/code/grid_search

# Do a GridSearch on two classifiers and compare them
from sklearn.model_selection import GridSearchCV
# could add "kernel": ["linear", "poly", "rbf"] below to try
rf_params = {'n_estimators': [50, 150, 250, 500],
               'min_samples_split': [2, 4, 6, 10, 20, 50]}

grid = GridSearchCV(rf, rf_params, cv=5, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
# find the best parameters and score for this dataset
print(f"Best Parameters: {grid.best_params_}")
print(f"Best Score: {grid.best_score_}")
print(f"Test Accuracy: {grid.score(X_test_scaled, y_test)}")
```
