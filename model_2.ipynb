# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("Resources/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]

# selecting all features except our y-value
selected_features = df.drop("koi_disposition", axis=1)
selected_features
```

```python
# target = df["koi_disposition"]
# target_names = ["CONFIRMED", "FALSE_POSITIVE"]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# set our X and y 
X = selected_features
y = df['koi_disposition']
print(X.shape, y.shape)
```

```python
# format the FALSE POSITIVE value 
mask = df["koi_disposition"] == "FALSE POSITIVE"
df.loc[mask, "koi_disposition"] = "FALSE_POSITIVE"
df["koi_disposition"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from tensorflow.keras.utils import to_categorical
```

```python
# MinMax Scaler 
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# label encoding
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
print(label_encoder.classes_)
encoded_y_train
```

```python
# to_categorical 
# y_train_categorical = to_categorical(encoded_y_train)
# print(y_train_categorical)
# y_train_categorical.dtype
# y_train_categorical.reshape(-1, 1)

# one-hot encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
y_train_categorical.shape
```

## Create Model Using Deep Learning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# from sklearn.svm import SVC
# model = SVC(kernel='linear')
# model.fit(X_train, y_train)
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=40))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```
